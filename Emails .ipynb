```python
#Computing list of All unique emails in the dataset.
all_emails = []
BCClist = []
for index, row in ef.iterrows():
    
    all_emails.append(row['Sender'])
    
    if not pd.isnull(row['To']):
        if ';' in row['To']:
            r = row['To'].split(';')
            for i in r:
                all_emails.append(i)
        else:
            all_emails.append(row['To'])
            
            
    if not pd.isnull(row['CC']):
        if ';' in row['CC']:
            r = row['CC'].split(';')
            for i in r:
                all_emails.append(i)
        else:
            all_emails.append(row['CC'])
        
        
    if not pd.isnull(row['BCC']):
        if ';' in row['BCC']:
            r = row['BCC'].split(';')
            for i in r:
                all_emails.append(i)
                BCClist.append(i)
        else:
            all_emails.append(row['BCC'])
            BCClist.append(i)
    
all_emails = list(set(all_emails))
len(all_emails)
```

```python
#Finding Communities By Department Names
departments = []
for i in all_emails:
    domains.append(i.split('@')[-1])
    if '-' in i:
        i = i.split('-')[-1]
        departments.append(i.split('<')[0])
        
domains = list(set(domains))
departments = list(set(departments))
```

```python
d = [item.replace(" ","").replace('"',"") for item in departments if len(item.replace(" ",""))==3 and item.isupper()]
d = [item.replace(")","").replace("'","") for item in d]
d = list(set(d))
d
```

```python
#Saving Communities list to csv
import csv
comm = {}
depscomm = {}
for deps in d:
    com = [x for i, x in enumerate(all_emails) if "- "+deps in x]
    if len(com) > 2: 
        depscomm[deps] = len(com)
        comm[deps] = com
        with open('Community.csv', 'a') as csv_file:
            writer = csv.writer(csv_file)
            for key, value in comm.items():
                writer.writerow([key, value])
```
